# Has life expectancy increased over time in the six nations?

```python
plt.figure(figsize=(10,6))
sns.lineplot(x = life_gdp_df.Year, y = life_gdp_df.LEAB, hue=life_gdp_df.Country)
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5), fontsize = 16)
plt.xlabel('Year')
plt.ylabel("Life expectancy at birth (years)")
plt.show()
```

You can see from the graph above that life expectancy has increased slow but steadily for five of the six countries. 
On the other hand, Zimbabwe saw a much more rapid increase over the years compared to the others.

```python
Leab_by_country = sns.FacetGrid(life_gdp_df, col="Country", col_wrap = 3, hue = "Country", sharey = False)
Leab_by_country = Leab_by_country.map(sns.lineplot,"Year","LEAB").set_axis_labels("Year","Life expectancy at birth (years)")
```

This shows each countries individual increase in life expectancy over the last 15 years. While a few of the countries saw dips in certain intervals, overall life expectancy has increased over all these countries.

# Has GDP increased over time in the six nations?



```python
plt.figure(figsize=(10,6))
sns.lineplot(x = life_gdp_df.Year, y = life_gdp_df.GDP, hue=life_gdp_df.Country)
plt.legend(loc='center right', bbox_to_anchor=(1.5, 0.5), fontsize = 16)
plt.xlabel('Year')
plt.ylabel("GDP (in trillions)")
plt.show()
```

This graphs all the countries GDP on one graph. You can see an increase in USA and China but it is difficult to see the other countries

```python
gdp_by_country = sns.FacetGrid(life_gdp_df, col="Country", col_wrap = 3, hue = "Country", sharey = False)
gdp_by_country = gdp_by_country.map(sns.lineplot,"Year","GDP").set_axis_labels("Year","GDP (in trillions)")
```

This graphs the GDP of each country seperately so you can get a better visual of the changes over time. You can clearly see that although some of the countries had dips in certain years overall they have all seen an increase in GDP.

# Is there a correlation between GDP and life expectancy of a country?
